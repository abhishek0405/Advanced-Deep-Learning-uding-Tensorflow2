```python
import tensorflow
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from tensorflow.keras.layers import Dense,Dropout,Conv2D,MaxPooling2D,Flatten,Input,concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical #for one hot encoding
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
img_size = x_train.shape[1] #height or width
x_train = np.reshape(x_train,[-1,img_size,img_size,1]) #img size and 1 as only one channel
x_test = np.reshape(x_test,[-1,img_size,img_size,1])
x_train=x_train/255
x_test=x_test/255
```

```python
#Building Y network CNN Model
left_inputs = Input(shape=(img_size,img_size,1))
filters=32
x = left_inputs
for i in range(3):
    x = Conv2D(filters,3,activation='relu',padding='same')(x)
    x = Dropout(0.3)(x)
    x = MaxPooling2D()(x)
    filters*=2
    
right_inputs=Input(shape=(img_size,img_size,1))
y=right_inputs
filters=32
for i in range(3):
    y = Conv2D(filters,3,activation='relu',padding='same',dilation_rate=2)(y)
    y = Dropout(0.3)(y)
    y = MaxPooling2D()(y)
    filters*=2
    
y=concatenate([x,y])
y = Flatten()(y)
y=Dropout(0.3)(y)
op = Dense(10,activation='softmax')(y)

```

```python
model = Model(inputs=[left_inputs,right_inputs],outputs=op)
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.load_weights('Y_CNN_MNIST.h5')
```

```python
model.fit([x_train,x_train],y_train,validation_data=([x_test,x_test],y_test),epochs=2,batch_size=32)
```

```python
model.evaluate([x_test,x_test],y_test)
```
